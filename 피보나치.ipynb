```python
import matplotlib.pyplot as plt
import numpy as np
from random import *
real_part = [] #실수 범위
imaginary_part = [] #허수 범우
x = [] # x 축

side = 10 # x 범위

def fibonacci(n): # 피보함수
    sqrt_5 = (5 ** 0.5)
    phi = (1 + sqrt_5) / 2
    psi = (1 - sqrt_5) / 2
    return (phi**n - psi**n) / sqrt_5

def plot_fibonacci_sequence(n): # 피보나치 계산
    fib = fibonacci(n)
    real_part.append(fib.real)
    imaginary_part.append(fib.imag)


t = int(input())
for __tests__ in range(t):
    #n = float(input("n을 입력하세요: "))
    #n = side*__tests__/t
    n = -side + 2*side*__tests__/t
    x.append(n)
    plot_fibonacci_sequence(n)


plt.plot(x, real_part, marker='o', markersize=1)
plt.xticks(np.arange(-side-1, side+1, 1))
plt.xlabel('N')
plt.ylabel('Real')
plt.title('Fibonacci Sequence in the Complex Plane')
plt.grid()
plt.show()


p = randint(2*t//side, t)


# 푸리에 변환 수행
fft_result = np.fft.fft(real_part)
frequencies = np.fft.fftfreq(t, d=(x[1] - x[0]))

# 결과를 시각화
plt.plot(frequencies, np.abs(fft_result), marker='o', markersize=1)
plt.title('Fourier')
plt.xlabel('Hz')
plt.ylabel('result')
plt.show()
```
